## Dissimilarity

Dalam peambangan data yang pengaplikasian yang sering digunakan adalah clustering. pada teknik ini perlu untuk menemukan seberapa mirip atau tidaknya data. Salah satu cara yaitu dengan menggunakan Dismilarity

import modul

```python
import pandas as pd
import math
```

Get Dataset

```python
df = pd.read_csv("https://raw.githubusercontent.com/Wing-Dimas/datamining/main/House_Rent_Dataset.csv")
```

```python
df.head()
```

ambil sample data

```python
data_sample = df[["Rent", "Size", "Area Type", "City", "Furnishing Status"]][0:2048:512].reset_index(drop=True)
data_sample
```

## Mencari similarity data normal

```python
def d1(i,j):
    P = 3
    M = 0
    for col in ["Area Type", "City", "Furnishing Status"]:
        if data_sample[col][i] == data_sample[col][j]:
            M += 1

    return (P-M)/P
```

```python
print(d1(0,1))
print(d1(0,2))
print(d1(0,3))
```

## Mencari similarity data numerik

```python
def d2(i,j, h):
    result = 0
    for col in ["Rent", "Size"]:
        result = math.pow(abs(data_sample[col][i] - data_sample[col][j]), h)
    return result**(1/h)
```

Manhattan distance

```python
print(d2(0,1,1))
print(d2(0,2,1))
print(d2(0,3,1))
```

Euclidean distance

```python
print(d2(0,1,2))
print(d2(0,2,2))
print(d2(0,3,2))
```
